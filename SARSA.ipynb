# SARSA

```python
gamma = 0.99
alpha = 0.5
epsilon = 0.1
episodes = 500
```

```python
# Q-table => store Q-Values

Q = np.zeros((48,4))
```

```python
# Policy - epsilon-greedy : state -> action

def epsilon_greedy(state):
  if random.random() < epsilon:
    return env.action_space.sample()  # random action => explore

  else:
    return np.argmax(Q[state])       # best action => exploit
  
```

```python
env = gym.make("CliffWalking-v1")

for episode in range(episodes):
    render = (episode % 50 == 0)

    if render:
        env.close()
        env = gym.make("CliffWalking-v1", render_mode="human")
        print(f"rendering env for episode = {episode+1}/{episodes}")

    state, _ = env.reset()
    action = epsilon_greedy(state)

    done = False
    total_reward = 0
    episode_len = 0

    while not done:
        next_state, reward, terminated, truncated, _ = env.step(action)
        done = terminated or truncated

        next_action = epsilon_greedy(next_state)

        # SARSA update
        Q[state, action] += alpha * (
            reward + gamma * Q[next_state, next_action] - Q[state, action]
        )

        state = next_state
        action = next_action

        total_reward += reward
        episode_len += 1

    print(f"episode = {episode+1}/{episodes}: total reward = {total_reward}, length = {episode_len}")

env.close()
```
